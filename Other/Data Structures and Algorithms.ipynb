```python
def binarySearch(cards,query):
    if len(cards) == 0:
        return -1
    elif len(cards)%2 == 0:
        firstCheck = int(len(cards)/2)
    else:
        firstCheck = int((len(cards)-1)/2)
        
    check = firstCheck
    lowerBound = 0
    upperBound = len(cards)-1
    
    while lowerBound<=upperBound: # check >= lowerBound and check <= upperBound ?
        if cards[check] == query:
            i=1
            if check != 0:
                while cards[check-i] == query:
                    i+=1
            return check - i + 1
        elif cards[check] > query:
            lowerBound = check+1
        else: # If cards[check] < query
            upperBound = check-1
        check = int((upperBound+lowerBound)/2)
    # Only continues if query not in cards
    return -1
```

Problem: find minimum number of rotations to get from a sorted list in ascending order to the given list.
Can assume that all elements are unique.
Rotation = replacement of final element in front of first element.

Input: List. Output: integer (number of rotations, or equivalently, index of minimum value)
Possible cases:
1. Minimum is in the middle.
2. Minimum is at the start.
3. Min is at the end.

Consider:

1. Elements may not be integers.
2. List may be empty. Return 0 if so.
3. List may just have 1 element. Return 0 if so.

```python
tests2 = []
tests2.append({
    "input":{"List":[1,2,3,4,5]},
    "output":0
})
tests2.append({
    "input":{"List":[3,4,5,1,2]},
    "output":3
})
tests2.append({
    "input":{"List":[2,3,4,5,1]},
    "output":4
})
tests2.append({
    "input":{"List":[]},
    "output":0
})
tests2.append({
    "input":{"List":[2.2,1.1]},
    "output":1
})
tests2.append({
    "input":{"List":[1]},
    "output":0
})
```

```python
# Brute force: this is O(n)

def getMinNumberOfRotations1(List):
    if len(List) == 0:
        return 0
    for i in range(len(List)):
        increased = List[i] > List[i-1] # If i = 0, compares to that List[-1], the last element, as desired.
        if (not increased):
            return i
```

```python
testAlgo(tests2,getMinNumberOfRotations1)
```

```python
# Aiming for O(log n)

def getMinNumberOfRotations2(List):
    if len(List)==0:
        return 0
    low = 0
    high = len(List)-1
    while low <= high:
        check = int((low+high)/2)
        increased = List[check] > List[check-1] # If check = 0, compares to List[-1], the last element, as desired.
        Next = check+1
        if Next == len(List):
            Next = 0
        increases = List[check] < List[Next]  
        if (not increased):
            return check    # If len(List)=1, arrive here (both increased and increases are false)
        elif (not increases):
            return Next
        elif List[check]<List[low]:
            high = check-1
        else:
            low = check+1 
```

Tree depth is log(n+1) <= log(n)+1.

Traversals of binary trees: preorder (root left right), in-order (left root right), postorder (left right root). Recurse on each node. Eventually you end up with a list of the node values with no repeats (unless multiple nodes have the same values).

```python
def traverse_preorder(tree):
    if tree is None:
        return []
    return [tree.key]+traverse_preorder(tree.left)+traverse_preorder(tree.right)
```

```python
def traverse_in_order(tree):
    if tree is None:
        return []
    return traverse_in_order(tree.left)+[tree.key]+traverse_in_order(tree.right)
```

```python
def traverse_postorder(tree):
    if tree is None:
        return []
    return traverse_postorder(tree.left)+traverse_postorder(tree.right)+[tree.key]
```

```python
# Make it possible to access all of these characterizations of a tree through the class, good programing practice.

class TreeNode:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
        
    def height(self):
        if self is None:
            return 0
        return 1 + max([TreeNode.height(self.left),TreeNode.height(self.right)])
    
    def size(self):
        if self is None:
            return 0
        return 1 + TreeNode.size(self.left) + TreeNode.size(self.right)
        
    def traverse_in_order(self):
        if self is None:
            return []
        return TreeNode.traverse_in_order(self.left)+[self.key]+TreeNode.traverse_in_order(self.right)
        
    def to_tuple(self):
        if self is None:
            return None

        mid = self.key
        left = tree_to_tuple(self.left)
        right = tree_to_tuple(self.right)

        if left is None and right is None:
            return mid

        return (left,mid,right)
    
    def display_keys(self,space="\t",level=0):
        if self is None:
            print(space*level+"N/A")
            return
        if self.left is None and self.right is None:
            print(space*level+str(self.key))
            return
        TreeNode.display_keys(self.right,space,level+1)
        print(space*level+str(self.key))
        TreeNode.display_keys(self.left,space,level+1)
        
    def __repr__(self):
        return f"Binary Tree <{self.to_tuple()}>"
        
    @staticmethod
    def parse_tuple(tree_tuple):
        if isinstance(tree_tuple,tuple) and len(tree_tuple)==3:
            node = TreeNode(tree_tuple[1])
            node.left = parse_tuple(tree_tuple[0])    # Could be nothing, or a node, or a subtree
            node.right = parse_tuple(tree_tuple[2])   # Could be nothing, or a node, or a subtree
        # Terminating conditions:
        elif tree_tuple is None:
            node = None
        else:
            node = TreeNode(tree_tuple)
        return node
```

Binary tree is a Binary Search Tree iff for every node, every node in the node's left subtree is less
than the node, and every node in the ode's right subtree is greater than the node. Every subtree of a BST is a BST.

```python
# Determine if a BT is a BST

def remove_none(List):
    return [x for x in List if x is not None]

def is_bst(node):   # Will return boolean based on BST-ness, max value, and min value in tree
    if node is None:
        return True, None, None
    
    is_bst_l,min_l,max_l = is_bst(node.left)
    is_bst_r,min_r,max_r = is_bst(node.right)
    
    is_bst_node = (is_bst_l and is_bst_r and (max_l is None or max_l<node.key) and (min_r is None or min_r>node.key))
    
    min_key = min(remove_none([min_l,node.key,min_r]))
    max_key = max(remove_none([max_l,node.key,max_r]))
    
    return is_bst_node,min_key,max_key
```

```python
def insert(node,key,value):   # Go left if smaller than root node, right if larger. Recurse on the subtree.
    if node is None:
        node = BSTNode(key,value)
    elif key < node.key:
        node.left = insert(node.left,key,value)
        node.left.parent = node
    elif key > node.key:
        node.right = insert(node.right,key,value)
        node.right.parent = node        
    return node
```

```python
def find(node,key):
    if node == None:
        return None
    if node.key == key:
        return node
    if node.key > key:
        return find(node.left,key)
    if node.key < key:
        return find(node.right,key)
```

```python
# Fxn to determine if BST is balanced. Left subtree balanced, right subtree balanced, diff. in subtree heights <= 1.
def is_balanced(tree):
    if tree is None:
        return True,0
    balanced_l,height_l = is_balanced(tree.left)
    balanced_r,height_r = is_balanced(tree.right)
    height = 1 + max([height_l,height_r])
    balanced = balanced_l and balanced_r and abs(height_l-height_r)<=1
    return balanced,height
```

```python
# Create balanced BST from sorted list of key-value pairs (list of tuples)
# Put middle element as root node. Break rest of list into left and right subtree lists. Recurse.
# Terminating conditions: [] returns None
def make_balanced_BST(data):
    if len(data)==0:
        return None
    n = int(len(data)/2) # Rounds down, so works correctly for odd numbers. Either way works for evens
    node = BSTNode(data[n][0],data[n][1])
    node.left = make_balanced_BST(data[:n])
    if node.left is not None:
        node.left.parent = node
    node.right = make_balanced_BST(data[n+1:])
    if node.right is not None:
        node.right.parent = node
    return node
```

```python
# Only had this in TreeNode class before, need it outside as well.
def size(tree):
    if tree is None:
        return 0
    return 1 + size(tree.left) + size(tree.right)
```

```python
class TreeMap():
    def __init__(self):
        self.root = None # Will contain BST object
        
    def __setitem__(self,key,value):  # Combines insert and update
        node = find(self.root,key) # Boolean, for if the given key is in the BST so far. True iff node is not None.
        if not node:
            self.root = insert(self.root,key,value)
            self.root = balance_BST(self.root)
        else:
            update(self.root,key,value)
    
    def __getitem__(self,key):
        node = find(self.root,key)
        return node.value if node else None
    
    # The above two methods allow you to set, update, and access values exactly like you do in dictionaries
    
    def __iter__(self):
        return (x for x in list_all(self.root)) # This is a generator, see use below
    
    def __len__(self):
        return size(self.root)
    
    def display(self):
        return display_keys(self.root)
```

### New Problem
Create hash table that supports:
1. Insert new key-value pair
2. Find value associated with key
3. Update value of a given key
4. List all keys

```python
class HashTable:
    def __init__(self,max_size=4096,hashFxn=get_valid_index):
        self.data_list=[None]*max_size # Will be a list of tuples, each tuple a key-value pair
        self.hashFxn = hashFxn
        
    def insert(self,key,value):
        index = self.hashFxn(key,self.data_list)
        self.data_list[index]=(key,value)
    
    def find(self,key):
        index = self.hashFxn(key,self.data_list)
        if self.data_list[index] is not None:
            value = self.data_list[index][1]
            return value
        else:
            return None
    
    def update(self,key,value):
        index = self.hashFxn(key,self.data_list)
        self.data_list[index]=(key,value)
    
    def list_all(self):
        return [KV[0] for KV in self.data_list if KV is not None]
```

```python
def basicHashFxn(key:str,data_list):
    max_index = len(data_list)
    index = 0
    for char in key:
        index += ord(char) # ord is a built-in fxn to turn characters into integers. ord("d") = 100, for example.
    index = index % max_index
    return index
```

```python
def get_valid_index(key:str,data_list,hashFxn=basicHashFxn):
    index = hashFxn(key,data_list)
    while data_list[index] is not None and data_list[index][0] is not key:
        index += 1
        if index == len(data_list):
            index = 0
    return index
```
